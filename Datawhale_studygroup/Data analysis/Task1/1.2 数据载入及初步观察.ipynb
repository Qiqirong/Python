Pandas使用：
Series是类似于Numpy中一组数组的对象，由一组任意类型的数据以及一组与之相关的数据标签（索引）组成。For example：

```python
S=pd.Series ([1,2,3,4], index=['a', 'b', 'c', 'd'])
print (S)
```

Another example:

```python
S1=pd.Series ({'pencil':2, 'pen':3,'notebook':4})
S2=pd.Series({'pen':8, 'pencil':10,'notebook':22})
print (S1+S2)
```

Series运算：
+: add()  e.g.: s1.add(s2, fill_value=0)
-: sub()  e.g.: s1.sub(s2, fill_value=0)
x：mul()  

Series 是一维数据，而 DataFrame 是二维数据。For example:

```python
data={
    'pencil': [2,4],
    'pen': [6,8],
    'notebook':[10,8]
}
df=pd.DataFrame(data, index=['20200819', '20200820'])
print (df)
```
